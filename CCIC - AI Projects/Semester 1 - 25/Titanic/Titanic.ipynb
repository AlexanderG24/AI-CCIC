Question: How many 1st class females were the lone survivor in their cabin?

```python
import pandas as pd

# Load dataset
df = pd.read_csv('titanic.csv')

# Count how many people shared each cabin
df['CabinCount'] = df.groupby('Cabin')['PassengerId'].transform('count')

# Filter for 1st class females who survived and were alone in their cabin
lone_survivors = df[(df['Pclass'] == 1) & 
                    (df['Sex'] == 'female') & 
                    (df['Survived'] == 1) & 
                    (df['CabinCount'] == 1)]

# Display the count
print("Number of 1st class female lone survivors in their cabin:", len(lone_survivors))
# Print their names
print(lone_survivors['Name'])
```
