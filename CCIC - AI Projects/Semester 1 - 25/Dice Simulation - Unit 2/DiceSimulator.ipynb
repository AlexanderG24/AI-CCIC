```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

# Step 1: Roll a 6-sided die 100 times
def roll_die(sides=6, trials=100):
    return [random.randint(1, sides) for _ in range(trials)]
    
results = roll_die(sides=6, trials=100)

# Step 2: Visualization with Seaborn
sns.set(style="whitegrid")
sns.histplot(results, bins=range(1, 8), discrete=True, kde=False)

plt.title("Dice Simulator (100 Rolls of a 6-Sided Die)")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.show()

# Roll a 12-sided die 200 times
results = roll_die(sides=12, trials=200)

sns.histplot(results, bins=range(1, 14), discrete=True, kde=False)
plt.title("Dice Simulator (200 Rolls of a 12-Sided Die)")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.show()
```
