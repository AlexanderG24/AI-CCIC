Question: In Monopoly, what is the theoretical probability of rolling either a 7 or an 11 on your first turn, and how does this compare to the empirical probability observed when simulating dice rolls?

Study:
I will conduct a probability experiment by simulating single-turn dice rolls. On each trial, I will roll two fair six-sided dice once and record the sum. The sample space for a single roll includes all possible sums from 2 through 12. I will check whether the sum is either 7 or 11; if so, I will count that trial as a success.

The theoretical probability can be calculated using the addition rule. The chance of rolling a 7 is 6 out of 36, and the chance of rolling an 11 is 2 out of 36. Since these events cannot occur at the same time, their probabilities are added together. This gives a combined probability of 8 out of 36, which reduces to 2 out of 9, or about 0.222.

To find the empirical probability, I will repeat the trial many times and divide the number of successful rolls (where the sum is 7 or 11) by the total number of trials. By comparing this empirical probability to the theoretical probability, I can see how closely the simulation matches what is expected in theory.

This experiment illustrates how the addition rule can be used to calculate the probability of either one event or another occurring, rather than just a single outcome. It also demonstrates that while rolling a 7 is the most common sum on two dice, including 11 increases the overall chance of success on a turn.

```python
import random
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set(style="whitegrid")

# Parameters
num_trials = 10000 

# Theoretical probabilities
p_7 = 6 / 36
p_11 = 2 / 36
p_7_or_11 = p_7 + p_11  # addition rule since mutually exclusive

print("Theoretical P(7) =", p_7)
print("Theoretical P(11) =", p_11)
print("Theoretical P(7 or 11) =", p_7_or_11, " (which is 2/9 ≈ 0.222222...)")

# Run simulation
results = []
success_count = 0
for _ in range(num_trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total = die1 + die2
    results.append(total)
    if total in (7, 11):
        success_count += 1

empirical = success_count / num_trials
print("\nNumber of successes (7 or 11):", success_count)
print("Empirical P(7 or 11) over", num_trials, "trials:", empirical)

# Also show empirical P(7) alone for comparison
count_7 = results.count(7)
empirical_7 = count_7 / num_trials
print("Empirical P(7) over", num_trials, "trials:", empirical_7)

# Visualization: distribution of sums
plt.figure(figsize=(10,6))
sns.countplot(x=results)
plt.title(f"Distribution of sums across {num_trials} trials")
plt.xlabel("Sum of two dice")
plt.ylabel("Frequency")
plt.show()
```

Conclusion:
The theoretical probability of rolling a 7 on a single turn is about 0.167, and the probability of rolling an 11 is about 0.056. Using the addition rule, the probability of rolling either a 7 or an 11 is approximately 0.222. In my simulation of 10,000 rolls, I observed 2,204 successes where the sum was 7 or 11, giving an empirical probability of 0.2204. For comparison, 1,673 rolls resulted in a 7 alone, yielding an empirical probability of 0.1673. These two pieces of quantitative evidence show that including 11 in addition to 7 increases the overall probability of success, and that the empirical results closely match the theoretical expectations.
