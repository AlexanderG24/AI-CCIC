Question: How many 1st class females were the lone survivor in their cabin?

This analysis is an observational study, we are not manipulating any variables or assigning passengers to conditions. Instead, we are observing and analyzing existing historical data from the Titanic passenger list to answer the question. The type of observational study we are performing is retrospective as we will collect data from the past. We will use a census because it is more exact and representative of everyone who survived. We have to look at all the people to see how many 1st class female's survived alone out of the whole population.

```python
import pandas as pd

# Load dataset
df = pd.read_csv('titanic.csv')
df.info()
# Count how many people shared each cabin
df['CabinCount'] = df.groupby('Cabin')['PassengerId'].transform('count')

# Filter for 1st class females who survived and were alone in their cabin
lone_survivors = df[(df['Pclass'] == 1) & 
                    (df['Sex'] == 'female') & 
                    (df['Survived'] == 1) & 
                    (df['CabinCount'] == 1)]

# Display the count
print()
print("Number of 1st class female lone survivors in their cabin:", len(lone_survivors))
print()
# Print their names
print(lone_survivors['Name'])

# 1. Determine how many samples (rows) are in the dataset
print()
print("Number of samples in dataset:", len(df))

# 2. Display the 7th sample (remember Python uses 0-based indexing, so 7th = index 6)
print("7th sample in dataset:")
print()
print(df.iloc[6])

# 3. Display the samples who embarked from Queenstown (Embarked = 'Q')
print()
queenstown_passengers = df[df['Embarked'] == 'Q']
print("Passengers who embarked from Queenstown:")
print(queenstown_passengers)

```
