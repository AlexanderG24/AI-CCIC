Question: What proportion of songs in Spotify’s Top Tracks USA playlist in 2024 came from Pop compared to all other genres combined?

The type of study I plan to conduct is an observational study because I am examining existing data about music genres without influencing or changing the outcomes. I will look at the entire population of songs in Spotify’s Top 50 USA playlist for 2024 in order to get the most accurate results.

The dataset was created by someone else, who manually recorded the genres from Spotify’s Top 50 playlist. An AI tool was also used to categorize and summarize the genres. A potential source of bias in this dataset is that the AI may overgeneralize or misclassify genres, leading to an inaccurate representation of certain categories. This could affect how precisely the data reflects the true distribution of music genres in the playlist.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Top Tracks Playlists - USA.csv")

# You can change the column name
genre_counts = df["AI Genre"].value_counts().reset_index()

# Rename columns for clarity
genre_counts.columns = ["Genre", "Count"]

# Display the frequency table
print(genre_counts)

# Plot a bar chart of genre frequency
plt.figure(figsize=(10,6))
sns.barplot(data=genre_counts, x="Genre", y="Count", palette="viridis")

plt.xticks(rotation=45, ha="right")
plt.title("Frequency of Music Genres in 2024 Top Tracks (USA - Spotify Top 50)")
plt.xlabel("Music Genre")
plt.ylabel("Number of Tracks")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Top Tracks Playlists - USA.csv")

# Replace with your actual column name
column_name = "AI Genre"  

# Create relative frequency table
relative_freq = df[column_name].value_counts(normalize=True).reset_index()

# Rename columns for clarity
relative_freq.columns = ["Category", "Relative Frequency"]

# Convert to percentages
relative_freq["Relative Frequency (%)"] = relative_freq["Relative Frequency"] * 100

# Display relative frequency table
print(relative_freq)

# Visualization
plt.figure(figsize=(10,6))
sns.barplot(data=relative_freq, x="Category", y="Relative Frequency (%)", palette="viridis")

plt.xticks(rotation=45, ha="right")
plt.title(f"Relative Frequency of {column_name} in Spotify US Top 50 (2024)")
plt.xlabel(column_name)
plt.ylabel("Relative Frequency (%)")
plt.show()
```

In Spotify’s Top Tracks USA playlist of 2024, Pop songs made up 0.40 (40%) of all tracks, while all other genres combined accounted for 0.60 (60%). This shows that Pop was the single most common genre but still less than the total of all other genres combined.
